#1.Running Sum of 1d Array

```python
from typing import List
class Solution:
    def runningSum(self, nums: List[int]) -> List[int]:
        l=[]
        sum1=0
        for i in nums:
            sum1+=i
            l.append(sum1)
        return l
a=Solution()
result=a.runningSum([1,1,1,1,1])
print(result)
```

#2.Shuffle the Array

```python
from typing import List
class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        first=nums[:n]
        last=nums[n:]
        lst=[]
        for i in range(n):
            lst.append(first[i])
            lst.append(last[i])
        return lst
a=Solution()
result=a.shuffle([2,5,1,3,4,7],3)
print(result)
```

#3.Kids With the Greatest Number of Candies

```python
from typing import List
class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        max1=max(candies)
        lst=[]
        for i in candies:
            sum1=i+extraCandies
            if sum1>=max1:
                lst.append(True)
            else:
                lst.append(False)
        return lst
a=Solution()
result=a.kidsWithCandies([2,3,5,1,3],3)
print(result)
```
