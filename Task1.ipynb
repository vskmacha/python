#2.Python if-else program

```python
n = int(input().strip())
if n%2==1 or (n%2==0 and 6<=n<=20):
    print("Weird")
elif (n%2==0 and 2<=n<=5) or n>20:
    print("Not Weird")
```

#6.write a function

```python
def is_leap(year):
    leap = False
    if year%400==0:
        leap=True
    elif year%100==0:
        leap=False
    elif year%4==0:
        leap=True
    else:
        leap=False
    return leap

year = int(input())
print(is_leap(year))
```
